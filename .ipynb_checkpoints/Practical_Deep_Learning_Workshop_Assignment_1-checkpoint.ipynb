**Information on the dataset from Wikipedia:**

The CIFAR-10 dataset (Canadian Institute For Advanced Research) is a collection of images that are commonly used to train machine learning and computer vision algorithms. It is one of the most widely used datasets for machine learning research.[1][2] The CIFAR-10 dataset contains 60,000 32x32 color images in 10 different classes.[3] The 10 different classes represent airplanes, cars, birds, cats, deer, dogs, frogs, horses, ships, and trucks. There are 6,000 images of each class.[4]

Computer algorithms for recognizing objects in photos often learn by example. CIFAR-10 is a set of images that can be used to teach a computer how to recognize objects. Since the images in CIFAR-10 are low-resolution (32x32), this dataset can allow researchers to quickly try different algorithms to see what works. Various kinds of convolutional neural networks tend to be the best at recognizing the images in CIFAR-10.

CIFAR-10 is a labeled subset of the 80 million tiny images dataset. When the dataset was created, students were paid to label all of the images.[5]

```
def show_samples(samples):
  fig = plt.figure(figsize=(12,12))
  position = 0
  for i in samples:
      ax = fig.add_subplot(2, 5, 1 + position, xticks=[], yticks=[])
      idx = np.where(y_train[:]==i)[0]
      features_idx = x_train[idx,::]
      img_num = np.random.randint(features_idx.shape[0])
      im = np.transpose(features_idx[img_num, ::], (0, 1, 2))
      ax.set_title(LABLES_DICT[i])
      position = position + 1
      plt.imshow(im)
  plt.show()
show_samples(range(len(LABLES_DICT)))
```

```
def plot_multiple_imgs(X,y,nrow=2,ncol=2,figsize=(13,7),preds=None,skip=0):
    fig,ax = plt.subplots(nrows=nrow,ncols=ncol,figsize=figsize)
    fig.subplots_adjust(hspace=0.1, wspace=0.1)
    for i in range(nrow*ncol):
        ax[i//ncol,i%ncol].imshow(X[skip+i],cmap='binary')
        ax[i//ncol,i%ncol].set_xticks([])
        ax[i//ncol,i%ncol].set_yticks([])
        if preds is not None:
            ax[i//ncol,i%ncol].text(0.05, 0.1, str(LABLES_DICT[int(preds[skip+i])]), transform=ax[i//ncol,i%ncol].transAxes,
                                   color='green' if y[skip+i]==preds[skip+i] else 'red',weight='bold', fontsize=20)
            ax[i//ncol,i%ncol].text(0.05, 0.25, str(LABLES_DICT[int(y[skip+i])]), color='blue',transform=ax[i//ncol,i%ncol].transAxes,weight='bold', fontsize=20)
        else:
            ax[i//ncol,i%ncol].text(0.05, 0.1, str(LABLES_DICT[int(y[skip+i])]), color='white',transform=ax[i//ncol,i%ncol].transAxes,weight='bold', fontsize=20)
    plt.show()
```

```
def plot_model_history(model_history):
    fig, axs = plt.subplots(1,2,figsize=(15,5))
    # summarize history for accuracy
    axs[0].plot(range(1,len(model_history.history['acc'])+1),model_history.history['acc'])
    axs[0].plot(range(1,len(model_history.history['val_acc'])+1),model_history.history['val_acc'])
    axs[0].set_title('Model Accuracy')
    axs[0].set_ylabel('Accuracy')
    axs[0].set_xlabel('Epoch')
    axs[0].set_xticks(np.arange(1,len(model_history.history['acc'])+1),len(model_history.history['acc'])/10)
    axs[0].legend(['train', 'val'], loc='best')
    # summarize history for loss
    axs[1].plot(range(1,len(model_history.history['loss'])+1),model_history.history['loss'])
    axs[1].plot(range(1,len(model_history.history['val_loss'])+1),model_history.history['val_loss'])
    axs[1].set_title('Model Loss')
    axs[1].set_ylabel('Loss')
    axs[1].set_xlabel('Epoch')
    axs[1].set_xticks(np.arange(1,len(model_history.history['loss'])+1),len(model_history.history['loss'])/10)
    axs[1].legend(['train', 'val'], loc='best')
    plt.show()
```

```
def print_stats_after_run(x_test_input, model_info_input):
  test_prediction = model.predict_classes(x_test_input)
  plot_multiple_imgs(x_test_input,y_test,5,5,(20,20),test_prediction,skip=500)
  print('model accuracy on test set is: {}%'.format(accuracy_score(y_test, test_prediction)*100))
  sns.heatmap(xticklabels=list(LABLES_DICT.values()),yticklabels=list(LABLES_DICT.values()),data = confusion_matrix(y_test, test_prediction),cmap='Greens', annot=True, fmt='d')
  plt.xlabel('Prediction')
  plt.ylabel('True label')
  plt.title('CIFER-10 Convolutional model \n classification results on test set')
  plot_model_history(model_info_input)
  print ("Model took %0.2f seconds to train"%(end - start))
  print()
```

```
def add_salt_pepper_noise(X_imgs):
    # Need to produce a copy as to not modify the original image
    X_imgs_copy = X_imgs.copy()
    row, col, _ = X_imgs_copy[0].shape
    salt_vs_pepper = 0.2
    amount = 0.004
    num_salt = np.ceil(amount * X_imgs_copy[0].size * salt_vs_pepper)
    num_pepper = np.ceil(amount * X_imgs_copy[0].size * (1.0 - salt_vs_pepper))
    
    for X_img in X_imgs_copy:
        # Add Salt noise
        coords = [np.random.randint(0, i - 1, int(num_salt)) for i in X_img.shape]
        X_img[coords[0], coords[1], :] = 1

        # Add Pepper noise
        coords = [np.random.randint(0, i - 1, int(num_pepper)) for i in X_img.shape]
        X_img[coords[0], coords[1], :] = 0
    return X_imgs_copy
  
def convert_color_images_to_black_and_white(X_imgs):
  # Need to produce a copy as to not modify the original image
  X_imgs_copy = X_imgs.copy()


  for X_img in X_imgs_copy:
      # Add Salt noise
      r, g, b = X_img[:,:,0], X_img[:,:,1], X_img[:,:,2]
      gray = 0.2989 * r + 0.5870 * g + 0.1140 * b
      X_img[:,:,0] = gray
      X_img[:,:,1] = gray
      X_img[:,:,2] = gray
  return X_imgs_copy
```

###Second Architecture

```
####################################################
#                Model Architecture Take 2         # Added Batch Normalization
####################################################


# Define seed
seed_num = 512
np.random.seed(seed_num)
tf.random.set_random_seed(seed_num)


model = Sequential()

model.add(Conv2D(48,(2,2), activation='relu',input_shape=(32,32,3)))
model.add(BatchNormalization())
model.add(Conv2D(48,(2,2),activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.05))
model.add(MaxPool2D())

model.add(Conv2D(96,(2,2),activation='relu'))
model.add(BatchNormalization())
model.add(Conv2D(96,(2,2),activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.1))
model.add(MaxPool2D())

model.add(Conv2D(144,(2,2),activation='relu'))
model.add(BatchNormalization())
model.add(Conv2D(144,(2,2),activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(MaxPool2D())

model.add(Flatten())
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(10,activation='softmax'))
model.summary()
```

```
###########
# 1st try # validation_split 0.15 + BatchNormalizaion
###########

model.compile(loss='categorical_crossentropy',optimizer=Adam(),metrics=['accuracy'])
filepath="weights.best_0.15_val_split.hdf5"
checkpoint = ModelCheckpoint(filepath, monitor='val_acc', verbose=1, save_best_only=True, mode='max')
callbacks_list = [checkpoint]

start = time.time()
model_info = model.fit(x_train,to_categorical(y_train),validation_split=0.15,shuffle=True,epochs=28, callbacks=callbacks_list, verbose=0)
end = time.time()

print_stats_after_run(x_test, model_info)
```

```
###########
# 2nd try # validation_split 0.1 + BatchNormalizaion
###########

model.compile(loss='categorical_crossentropy',optimizer=Adam(),metrics=['accuracy'])
filepath="weights.best_0.1_val_split.hdf5"
checkpoint = ModelCheckpoint(filepath, monitor='val_acc', verbose=1, save_best_only=True, mode='max')
callbacks_list = [checkpoint]

start = time.time()
model_info = model.fit(x_train,to_categorical(y_train),validation_split=0.1,shuffle=True,epochs=28, callbacks=callbacks_list, verbose=0)
end = time.time()

print_stats_after_run(x_test, model_info)
```

```
###########
# 3rd try # validation_split 0.1 + BatchNormalizaion + Data Augmuntation
###########

filepath="weights.best_0.1_BN_AUG_NOLASTDROPOUT.hdf5"
checkpoint = ModelCheckpoint(filepath, monitor='val_acc', verbose=1, save_best_only=True, mode='max')
callbacks_list = [checkpoint]

x_train_augmented, x_test_augmented, y_train_augmented, y_test_augmented = train_test_split(x_train, y_train, test_size=0.1, random_state=seed_num)

dataAug = ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.05,
    height_shift_range=0.05,
    horizontal_flip=True,
    zoom_range = 0.1,
    )
dataAug.fit(x_train_augmented)

model.compile(loss='categorical_crossentropy',optimizer=Adam(),metrics=['accuracy'])

start = time.time()
model.fit_generator(dataAug.flow(x_train_augmented, to_categorical(y_train_augmented), batch_size=32,shuffle = True),
                    steps_per_epoch=len(x_train_augmented) / 32, epochs=28, validation_data = (x_test_augmented, to_categorical(y_test_augmented)), callbacks=callbacks_list)
end = time.time()
```

```
scores = model.evaluate(x_test, to_categorical(y_test))
print('\nTest result: %.2f "%" loss: %.3f' % (scores[1]*100,scores[0]))
```

#### c.	Try to figure out where & why is the model misclassifying and suggest at least 3 ways to improve the results

We figure out the model is misclassifying for a number of reasons:

1. The distance of the object in the images is inconsistent.  There are images of objects taken remotely, and images of objects taken close up. For example image of an airplane on the ground, as opposed to an image of a plane in the air.

2. The color of the object in the images is inconsistent. There are images of the same objects in different colors. For example, image of red car, as opposed to an image of yellow car.

3. The angle of the object in the images is inconsistent. There are images of the same objects in different angles. For example, image of back car, as opposed to an image of car from the side.

4. The part of the object in the images is inconsistent. There are images of objects that have not been taken fully.  For example, there are images of cats and dogs, whose only their head was taken, and images of cats and dogs, whose all their body was taken.


#### d.	Prioritize the list of suggestions for improvements and implement the first 2 suggestions and repeat section b.

We suggest ways to improve the results:
1. Add batch normalization layers, to normalize  the data after all activation layer.

2. Add augmentation, to ignore  the distances of the objects in the images, and to ignore the angles of the objects in the images.

3. Convert the 3 channels of RGB color model to one channel between 0 and 255, to ignore the colors of the images.

4. Adding Salt and Pepper noise, to ignore the quality images.

```
####################################################
#                Transfer
####################################################
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(32, 32, 3))
last = base_model.get_layer('block3_pool').output
x = Flatten()(last)
x = Dense(10, activation='relu')(x)
x = BatchNormalization()(x)
x = Dense(10, activation='softmax')(x)
x = Dropout(0.25)(x)

filepath="weights.best_0.1_val_split_VGG16.hdf5"
checkpoint = ModelCheckpoint(filepath, monitor='val_acc', verbose=1, save_best_only=True, mode='max')
callbacks_list = [checkpoint]

model = Model(base_model.input, x)
model.summary()

model.compile(loss='categorical_crossentropy',optimizer=Adam(),metrics=['accuracy'])
start = time.time()
model_info = model.fit(x_train,to_categorical(y_train),validation_split=0.1,shuffle=True,callbacks=callbacks_list,epochs=28)
end = time.time()
test_preiction = model.predict(x_test)
test_preiction = np.argmax(test_preiction,axis=1)
print('model accuracy on test set is: {}%'.format(accuracy_score(y_test,test_preiction)*100))
sns.heatmap(xticklabels=list(LABLES_DICT.values()),yticklabels=list(LABLES_DICT.values()),data = confusion_matrix(y_test, test_preiction),cmap='Greens', annot=True, fmt='d')
plt.xlabel('Prediction')
plt.ylabel('True label')
plt.title('CIFER-10 Convolutional model \n classification results on test set')
plot_model_history(model_info)
print ("Model took %0.2f seconds to train"%(end - start))
print()
```

```
####################################################
#                Feature Extraction
####################################################
# Define seed
seed_num = 512
np.random.seed(seed_num)
tf.random.set_random_seed(seed_num)

base_model = VGG16(weights='imagenet', include_top=False, input_shape=(32, 32, 3))
last = base_model.get_layer('block3_pool').output
x = BatchNormalization()(last)
x = Flatten()(x)
x = BatchNormalization()(x)
x = Dense(10, activation='relu')(x)

filepath="weights.best_0.1_val_split_VGG16_feature_extraction.hdf5"
checkpoint = ModelCheckpoint(filepath, monitor='val_acc', verbose=1, save_best_only=True, mode='max')
callbacks_list = [checkpoint]

model = Model(base_model.input, x)
model.summary()

model.compile(loss='categorical_crossentropy',optimizer=Adam(),metrics=['accuracy'])
start = time.time()
model_info = model.fit(x_train_org,to_categorical(y_train),validation_split=0.1,shuffle=True,callbacks = callbacks_list, epochs=50)
feature_x_train = model.predict(x_train_org)
feature_x_test = model.predict(x_test_org)
end = time.time()
```

```
####################################################
#                Feature Extraction by KNN
####################################################
neigh = KNeighborsClassifier(n_neighbors=10)
neigh.fit(feature_x_train, y_train.ravel()) 
test_prediction = neigh.predict(feature_x_test)
print('model accuracy on test set is: {}%'.format(accuracy_score(y_test,test_prediction)*100))
sns.heatmap(xticklabels=list(LABLES_DICT.values()),yticklabels=list(LABLES_DICT.values()),data = confusion_matrix(y_test, test_prediction),cmap='Greens', annot=True, fmt='d')
plt.xlabel('Prediction')
plt.ylabel('True label')
plt.title('CIFER-10 Convolutional model \n classification results on test set')
print ("Model took %0.2f seconds to train"%(end - start))
plot_multiple_imgs(x_test,y_test,5,5,(20,20),test_prediction,skip=500)
print()
```

```
####################################################
#     Feature Extraction by Logistic Regression    #
####################################################
lrc = LogisticRegression(n_jobs=8,solver='lbfgs')
lrc.fit(feature_x_train,y_train.ravel())
test_prediction = lrc.predict(feature_x_test)
print('model accuracy on test set is: {}%'.format(accuracy_score(y_test,test_prediction)*100))
sns.heatmap(xticklabels=list(LABLES_DICT.values()),yticklabels=list(LABLES_DICT.values()),data = confusion_matrix(y_test, test_prediction),cmap='Greens', annot=True, fmt='d')
plt.xlabel('Prediction')
plt.ylabel('True label')
plt.title('CIFER-10 Convolutional model \n classification results on test set')
print ("Model took %0.2f seconds to train"%(end - start))
plot_multiple_imgs(x_test,y_test,5,5,(20,20),test_prediction,skip=500)
print()
```
